```python
import pandas as pd

```

```python
import matplotlib.pyplot as plt
```

```python
import locale
```

```python
from matplotlib.ticker import FuncFormatter
```

```python
# Configurar locale para formatar valores em R$
locale.setlocale(locale.LC_ALL, 'pt_BR.UTF-8')
```

```python
#Dicionário de faturamento
dict_faturamento = {
   'data_ref': [
        '2023-01-01', 
        '2020-02-01', 
        '2021-03-01', 
        '2022-04-01', 
        '2023-05-01',
        '2023-06-01', 
        '2020-07-01', 
        '2021-08-01', 
        '2022-09-01', 
        '2023-10-01',
        '2022-11-01', 
        '2023-12-01',
        ],
    'valor': [
        400000, 
        890000, 
        760000, 
        430000, 
        920000,
        340000, 
        800000, 
        500000, 
        200000, 
        900000,
        570000, 
        995000,
        ]
}
```

```python
df_faturamento = pd.DataFrame(dict_faturamento)
```

```python
df_faturamento
```

## Calculo da média das vendas

```python
# Calculando a media das vendas
df_faturamento['valor'].mean()
```

```python
#Convertendo data_ref de string para datetime
df_faturamento['data_ref'] = pd.to_datetime(df_faturamento['data_ref'])
```

```python
# Grafico de Barras Vertical - Data (ordenado)
df_faturamento.sort_values(by='data_ref').plot.bar(x='data_ref', y='valor')
```

## Gráfico de barras vertical mostrando o mês de referência e o valor

```python
# Criar o gráfico de barras com ajuste de largura
fig, ax = plt.subplots(figsize=(10, 6))
bars = ax.bar(
    df_faturamento['data_ref'], df_faturamento['valor'],
    color='skyblue', edgecolor='black', width=10  # Ajuste da largura da barra
)

# Adicionar os valores reais no topo de cada barra com formato de dinheiro
for bar in bars:
    yval = bar.get_height()
    ax.text(
        bar.get_x() + bar.get_width() / 2,
        yval,
        locale.currency(yval, grouping=True),  # Formatar como R$ com separadores
        ha='center',
        va='bottom',
        fontsize=10,
        color='black'
    )

# Formatar o eixo Y como moeda (R$)
def format_currency(x, pos):
    return locale.currency(x, grouping=True)

ax.yaxis.set_major_formatter(FuncFormatter(format_currency))

# Personalizar o gráfico
ax.set_xlabel('Data de Referência', fontsize=12, labelpad=10)
ax.set_ylabel('Faturamento (R$)', fontsize=12, labelpad=10)
ax.set_title('Faturamento Mensal', fontsize=16, pad=15)
ax.grid(axis='y', linestyle='--', alpha=0.7)
plt.xticks(rotation=45, fontsize=10)
plt.yticks(fontsize=10)

# Ajustar espaçamento
plt.tight_layout()

# Exibir o gráfico
plt.show()
```

## Gráfico de linhas mostrando o mês de referência e o valor.

```python
#Dicionário de faturamento
dict_faturamento = {
   'data_ref': [
        '2023-01-01', 
        '2020-02-01', 
        '2021-03-01', 
        '2022-04-01', 
        '2023-05-01',
        '2023-06-01', 
        '2020-07-01', 
        '2021-08-01', 
        '2022-09-01', 
        '2023-10-01',
        '2022-11-01', 
        '2023-12-01',
        ],
    'valor': [
        400000, 
        890000, 
        760000, 
        430000, 
        920000,
        340000, 
        800000, 
        500000, 
        200000, 
        900000,
        570000, 
        995000,
        ]
}
df_faturamento = pd.DataFrame(dict_faturamento)

# Criar o gráfico de linha com layout ajustado
fig, ax = plt.subplots(figsize=(14, 8))  # Use constrained_layout
ax.plot(
    df_faturamento['data_ref'], df_faturamento['valor'],
    marker='o', color='blue', linewidth=2, markersize=8, label='Faturamento'
)

# Adicionar os valores reais no topo de cada ponto
for i, valor in enumerate(df_faturamento['valor']):
    ax.text(
        i, valor,
        locale.currency(valor, grouping=True),  # Formatar como R$ com separadores
        ha='center',
        va='bottom',
        fontsize=12,
        color='black'
    )

# Formatar o eixo Y como moeda (R$)
def format_currency(x, pos):
    return locale.currency(x, grouping=True)

ax.yaxis.set_major_formatter(FuncFormatter(format_currency))

# Personalizar o gráfico
ax.set_xlabel('Mês de Referência', fontsize=14, labelpad=15)
ax.set_ylabel('Faturamento (R$)', fontsize=14, labelpad=15)
ax.set_title('Faturamento Mensal', fontsize=18, pad=20)
ax.grid(axis='y', linestyle='--', alpha=0.7)
plt.xticks(rotation=45, fontsize=12)
plt.yticks(fontsize=12)
ax.legend(fontsize=12)

# Ajustar o espaçamento (se necessário)
plt.subplots_adjust(left=0.1, right=0.95, top=0.9, bottom=0.3)

# Exibir o gráfico
plt.show()
```
